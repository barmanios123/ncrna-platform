```python
import pandas as pd
import numpy as np

path = "data/raw/EBPlusPlusAdjustPANCAN_IlluminaHiSeq_RNASeqV2.geneExp.tsv"

target_lncrnas = [
    "HNF1A-AS1", "NEAT1", "MALAT1", "SNHG12", "HULC", "HOTAIR",
    "GAS5", "LINC01116", "MIAT", "XIST", "H19", "MEG3"
]

df = pd.read_csv(path, sep="\t")

gene_col = df.columns[0]
df = df.rename(columns={gene_col: "gene_symbol"})

df["gene_symbol"] = df["gene_symbol"].astype(str).str.strip()

expr = df[df["gene_symbol"].isin(target_lncrnas)].copy()

sample_cols = [c for c in expr.columns if c != "gene_symbol"]
expr[sample_cols] = expr[sample_cols].apply(pd.to_numeric, errors="coerce")

summary = pd.DataFrame({
    "gene_symbol": expr["gene_symbol"],
    "expr_mean_tcga_pancan": expr[sample_cols].mean(axis=1),
    "expr_median_tcga_pancan": expr[sample_cols].median(axis=1),
    "expr_std_tcga_pancan": expr[sample_cols].std(axis=1),
    "expr_min_tcga_pancan": expr[sample_cols].min(axis=1),
    "expr_max_tcga_pancan": expr[sample_cols].max(axis=1),
    "expr_q1_tcga_pancan": expr[sample_cols].quantile(0.25, axis=1),
    "expr_q3_tcga_pancan": expr[sample_cols].quantile(0.75, axis=1),
})

summary["expr_iqr_tcga_pancan"] = (
    summary["expr_q3_tcga_pancan"] - summary["expr_q1_tcga_pancan"]
)

summary["expr_prevalence_tcga_pancan"] = (
    (expr[sample_cols] > 0).sum(axis=1) / len(sample_cols)
)

summary["in_tcga_pancan_expr"] = True

summary = summary.drop_duplicates(subset=["gene_symbol"]).reset_index(drop=True)

summary.to_csv(
    "data/processed/tcga_pancan_expression_summary_features.csv",
    index=False
)

print(summary)
```

```python
import pandas as pd
import numpy as np

tcga_raw_path = "data/raw/EBPlusPlusAdjustPANCAN_IlluminaHiSeq_RNASeqV2.geneExp.tsv"
tcga_raw = pd.read_csv(tcga_raw_path, sep="\t")

print("Raw TCGA shape:", tcga_raw.shape)

# 1) Split gene_id into symbol and entrez
gene_split = tcga_raw["gene_id"].astype(str).str.split("|", n=1, expand=True)
tcga_raw["gene_symbol"] = gene_split[0].str.strip()
tcga_raw["entrez_id"] = gene_split[1].str.strip()

# 2) Identify sample columns
sample_cols = [c for c in tcga_raw.columns if c not in ["gene_id", "gene_symbol", "entrez_id"]]
print("Number of sample columns:", len(sample_cols))

num = tcga_raw[sample_cols].apply(pd.to_numeric, errors="coerce")

# 3) Aggregate across samples
tcga_pancan_df = pd.DataFrame({
    "gene_symbol": tcga_raw["gene_symbol"],
    "expr_mean_tcga_pancan": num.mean(axis=1),
    "expr_median_tcga_pancan": num.median(axis=1),
    "expr_std_tcga_pancan": num.std(axis=1),
    "expr_min_tcga_pancan": num.min(axis=1),
    "expr_max_tcga_pancan": num.max(axis=1),
    "expr_q1_tcga_pancan": num.quantile(0.25, axis=1),
    "expr_q3_tcga_pancan": num.quantile(0.75, axis=1),
})

tcga_pancan_df["expr_iqr_tcga_pancan"] = (
    tcga_pancan_df["expr_q3_tcga_pancan"] - tcga_pancan_df["expr_q1_tcga_pancan"]
)
tcga_pancan_df["expr_prevalence_tcga_pancan"] = (num > 0).mean(axis=1)
tcga_pancan_df["in_tcga_pancan_expr"] = (tcga_pancan_df["expr_prevalence_tcga_pancan"] > 0).astype(int)

print("TCGA pancan summary shape:", tcga_pancan_df.shape)
print(tcga_pancan_df.head())

tcga_pancan_df.to_csv(
    "data/processed/tcga_pancan_expression_summary_features.csv",
    index=False
)
```

```python
import sqlite3
import pandas as pd

tcga_cols = [
    "expr_mean_tcga_pancan",
    "expr_median_tcga_pancan",
    "expr_std_tcga_pancan",
    "expr_min_tcga_pancan",
    "expr_max_tcga_pancan",
    "expr_q1_tcga_pancan",
    "expr_q3_tcga_pancan",
    "expr_iqr_tcga_pancan",
    "expr_prevalence_tcga_pancan",
    "in_tcga_pancan_expr",
]

conn = sqlite3.connect("ncrna_platform.db")
cur = conn.cursor()

existing_cols = pd.read_sql_query("PRAGMA table_info(target_scores);", conn)["name"].tolist()

for col in tcga_cols:
    if col not in existing_cols:
        col_type = "INTEGER" if col == "in_tcga_pancan_expr" else "REAL"
        cur.execute(f"ALTER TABLE target_scores ADD COLUMN {col} {col_type}")

conn.commit()

update_df = df[["ncrna_id"] + tcga_cols].copy()

for _, r in update_df.iterrows():
    cur.execute(
        """
        UPDATE target_scores
        SET expr_mean_tcga_pancan = ?,
            expr_median_tcga_pancan = ?,
            expr_std_tcga_pancan = ?,
            expr_min_tcga_pancan = ?,
            expr_max_tcga_pancan = ?,
            expr_q1_tcga_pancan = ?,
            expr_q3_tcga_pancan = ?,
            expr_iqr_tcga_pancan = ?,
            expr_prevalence_tcga_pancan = ?,
            in_tcga_pancan_expr = ?
        WHERE ncrna_id = ?
        """,
        (
            float(r["expr_mean_tcga_pancan"]),
            float(r["expr_median_tcga_pancan"]),
            float(r["expr_std_tcga_pancan"]),
            float(r["expr_min_tcga_pancan"]),
            float(r["expr_max_tcga_pancan"]),
            float(r["expr_q1_tcga_pancan"]),
            float(r["expr_q3_tcga_pancan"]),
            float(r["expr_iqr_tcga_pancan"]),
            float(r["expr_prevalence_tcga_pancan"]),
            int(r["in_tcga_pancan_expr"]),
            r["ncrna_id"],
        ),
    )

conn.commit()

check = pd.read_sql_query(
    """
    SELECT
        ncrna_id,
        expr_mean_tcga_pancan,
        expr_median_tcga_pancan,
        in_tcga_pancan_expr
    FROM target_scores
    ORDER BY ncrna_id
    """,
    conn,
)

conn.close()

print(check.to_string(index=False))
```
